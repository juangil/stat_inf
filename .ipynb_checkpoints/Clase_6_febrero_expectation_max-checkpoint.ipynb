```python
% pylab inline
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from pylab import *
from matplotlib import cm
```

# Plotting the Log Likelihood

```python
N = 100
M = 1000
mu = np.linspace(0.0, 1.0, M)
var = np.linspace(0.5,2.0, M)
z = np.zeros((M,M))
XX = np.random.normal(0.0, 1.0, N)
for i in range(0, mu.shape[0]):
    for j in range(0, var.shape[0]):
        X = XX[:,None]
        mult = (np.transpose(X) - mu[i])
        mult2 = X - mu[i]
        remult = np.dot(mult,mult2)
        img = ((-1.0/(2*var[i]))*(remult)) - ((N/2.0)*np.log(math.pi*2.0*var[j]))
        z[i][j] = img
pb.imshow(z, extent=[0.0,1.0,0.5,2.0])
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
#ax.plot_trisurf(mu, var, np.diag(z), cmap=cm.jet, linewidth=0.2, vmin = 0.0, vmax = 2.0)
ax.contour(mu, var, z)#, cmap=cm.coolwarm)
```

# Expectation maximization

```python
N = 50
D = 2
M = 1
ini_z = np.random.normal(0.0,1.0, N)
W = np.random.random_sample((D,M))
ini_z = ini_z[:,None]
W[0] = 0.5
W[1] = 0.2
X = ini_z*np.transpose(W)
cov_noise = np.eye(D)*0.05
mean_noise = np.zeros(D)
sample_noise = np.random.multivariate_normal(mean_noise, cov_noise, N)
X = X + sample_noise
pb.plot(X[:,0], X[:,1], 'bo')
```

```python
Qold = 0.0
Qnew = 1.0
EPS = 0.00001
W_estim = np.random.random_sample((D,M))
W_estim = np.matrix(W_estim)
sigma_estim = 0.05
cont = 0
while(np.fabs(Qold - Qnew) >= EPS and cont <= 1000):
    # calcular Qold
    M_tmp = np.transpose(W_estim)*W_estim + sigma_estim*np.eye(M)
    exp_z = (np.linalg.inv(M_tmp)*np.transpose(W_estim))*np.transpose(X)
    cov_z = ((1.0/(sigma_estim))*np.linalg.inv(M_tmp)) + (exp_z*np.transpose(exp_z))
    # actualizar parametros
    sample_mean = np.sum(X, axis=0)
    sample_mean = sample_mean*(1/X.shape[0])
    sample_mean = sample_mean[:,None]
    #print X.shape, exp_z.shape, sample_mean.shape
    m1 = X
    for i in range(0, X.shape[0]):
        m1[i,:] = m1[i,:] - np.transpose(sample_mean)
    W_estim = (m1.T*exp_z.T)*np.linalg.inv(cov_z)
    cont += 1
```
